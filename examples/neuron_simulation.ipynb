```python
from snn.models import Neuron
from snn.input import RandomStimulator
```

## Initialze neuron and stimulator

```python
neuron = Neuron(
    rest_potential=0.0,
    min_potential = -1.0,
    threshold_potential = 25.0,
    excitatory_potential = 4.0,
    leakage_factor = 0.25,
    rest_period = 40
)

stimulator = RandomStimulator()
```

## Simulation

```python
total_time_step = 500
time = range(total_time_step)
input_potentials = []
neuron_potentials = []
output_spikes = []

for i in time:
  input_potential = stimulator()
  spike, neuron_potential = neuron(input_potential)

  input_potentials.append(input_potential)
  neuron_potentials.append(neuron_potential)
  output_spikes.append(int(spike))
```

```python
import matplotlib.pyplot as plt
```

## Input

```python
plt.plot(time, input_potentials)
plt.xlabel("timestep")
plt.ylabel("input potential")
plt.show()
```

## Neuron Potential

```python
plt.plot(time, neuron_potentials)
plt.xlabel("timestep")
plt.ylabel("neuron potential")
plt.show()
```

## Neuron Spike

```python
plt.plot(time, output_spikes)
plt.xlabel("timestep")
plt.ylabel("spikes")
plt.show()
```
